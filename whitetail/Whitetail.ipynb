```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


whitetail = pd.read_csv("whitetailweather.csv", parse_dates=["DATE"])

whitetail.drop(columns=["WT03", "WT06", "WT11", "WT01", "DAPR", "MDPR", "DASF", "MDSF"], inplace=True)

whitetail.rename(columns={"SNOW": "snow_fall", "WT04": "sleet", "PRCP": "precipitation", 
                         "WT05": "hail", "TOBS": "observed_temp",
                         "SNWD": "snow_depth", "TMAX": "max_temp", "WESD": "water_eq_snow",
                         "TMIN": "min_temp"}, inplace=True)
```

```python
snow = whitetail[whitetail["snow_depth"] > 0]

fig,ax = plt.subplots(figsize=(15,8))
sns.scatterplot(x="DATE", y="snow_depth", data=snow)
```
